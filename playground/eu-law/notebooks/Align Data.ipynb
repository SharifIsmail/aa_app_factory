# Align EUR-LEX data with ground truth

```python
import pandas as pd
filenames = ["../eur-lex/eur_lex_data_2023-10-01_to_2023-12-31.xlsx", 
             "../eur-lex/eur_lex_data_2024-01-01_to_2024-12-31.xlsx", 
             "../eur-lex/eur_lex_data_2025-01-01_to_2025-05-19.xlsx"]
eurlex = pd.concat([pd.read_excel(f) for f in filenames], ignore_index=True)
eurlex['merge_key'] = eurlex['expression_url'].str.rsplit('/', n=1).str[-1]
```

```python
eurlex
```

```python
ground_truth = pd.read_excel("consolidated-legal-acts.xlsx")
```

```python
ground_truth
```

```python
# enrich ground truth table with metadata
ground_truth.rename(columns={'oj_doc_id': 'merge_key'}, inplace=True)
merged_df = pd.merge(eurlex, ground_truth, on='merge_key', how='left')
```

```python
merged_df.to_excel('merged-list-eurlex-ground-truth.xlsx')
```
